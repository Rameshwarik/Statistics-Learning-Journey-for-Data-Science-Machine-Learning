## Calculating mean

### Method 1 - Manual calculation

```python
population = [150,160,170,165,155]
sample = [150,170,165]

pop_mean =sum(population)/ len(population)
sample_mean = sum(sample)/ len(sample)

print('Population mean is', pop_mean)
print('Sample mean is', sample_mean)
```

### Method 2 - Using Numpy library

```python
import numpy as np

data = [10,20,30,40,50]

arithmatic_mean = np.mean(data)

print('The mean is',arithmatic_mean)
```

### Method 3 - using statistics

```python
import statistics as stats

var = [15,30,45,60,75]

mean_stats = stats.mean(var)

print('The mean is',mean_stats)
```

## MODE

### Method 1 : statistics

```python
import statistics as stats

data = [10,20,10,10,20,30]

mode = stats.mode(data)
print('The Mode is', mode)
```

```python
import statistics as stats

data = [10,20,10,10,20,30,20,40]

mode = stats.multimode(data)
print('The Mode is', mode)
```

## Varience

### Method 1 : Using Numpy

```python
import numpy as np

data = [2,4,4,5,6,7,8,9]

sample_varience = np.var(data , ddof =1)

Population_varience = np.var(data , ddof =0)

print('The sample varience is',sample_varience)
print('The population varience is', Population_varience)
```

### Method 2 : Using pandas

```python
import pandas as pd

data = [4,5,6,78,9,4,5,6,7,8,5,2,4,32,65,1,5,25]
series = pd.Series(data)
print(series)

samp_var =  series.var()
print('The sample varience is',samp_var)

pop_var = series.var(ddof=0)
print('The population varience is', pop_var)
```

### Methode 3 : Using Scipy

```python
from scipy import stats

data = [7,8,9,4,5,6,1,2,3,12]

sam_var = stats.tvar(data)
print('The sample varience is',samp_var)

pop_var = stats.tvar(data, ddof = 0)
print('The population varience is', pop_var)
```

## Standard Deviation

### Using Numpy

```python
import numpy as np
data = [60,65,7,0,75,85,70,80]

sam_std = np.std(data,ddof=1)
print('The sample std deviation is',sam_std)

pop_std = np.std(data,ddof=0)
print('The population std deviation is',pop_std)
```

### Using Pandas

```python
import pandas as pd

data = [4,5,6,78,9,4,5,6,7,8,5,2,4,32,65,1,5,25]
series = pd.Series(data)
print(series)

samp_std =  series.std()
print('The sample std deviation is',samp_std)

pop_std = series.std(ddof=0)
print('The population std deviation is', pop_std)
```

### Using Scipy

```python
from scipy import stats
import math

data = [60,65,7,0,75,85,70,80]

sam_std = math.sqrt(stats.tvar(data))
print('The sample std deviation is',sam_std)

pop_std = math.sqrt(stats.tvar(data,ddof=0))
print('The population std deviation is',pop_std)
```
